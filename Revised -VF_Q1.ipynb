```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
import requests
import json
from datetime import datetime
```

```python
# Reading the final dataset NYbike_trips_Q3.csv

filepath = '/Users/vineetasinha/Documents/workspace/NYCitiBike/NYbike_trips_Q1.csv'
df_NYTrips_Q1 = pd.read_csv(filepath)
```

```python
# Count missing values per column
missing_values_count = df_NYTrips_Q1.isnull().sum()

# Display count of missing values
print("Missing values count per column:")
print(missing_values_count)
```

```python
# Cleaning the missing values
df_cleaned_q1 = df_NYTrips_Q1.dropna(subset=['AvgTemp','end_station_name','end_station_id'])
```

```python
missing_values_count = df_cleaned_q1.isnull().sum()

# Display count of missing values
print("Missing values count per column:")
print(missing_values_count)
```

```python
# converting the 'Date' column in the df_NYTrips_Q1 DataFrame to a datetime format. It includes an additional parameter errors='coerce', which handles any errors that may arise during the conversion.
df_cleaned_q1['Date'] = pd.to_datetime(df_cleaned_q1['Date'], errors='coerce')
```

```python
# creating a new DataFrame named daily_trips_q1 that contains the count of bike trips per day for the second quarter (Q2).
daily_trips_q1 = pd.DataFrame(df_cleaned_q1.groupby(['Date'])['ride_id'].count()).reset_index()
```

```python
# rename a column in the daily_trips_q2 from ride_id to bike_rides_daily
daily_trips_q1.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)

# merging two DataFrames (df_NYTrips_Q2 and daily_trips_q2) on a common column, 'Date', while renaming a column from daily_trips_q1 in the process. The merge is performed using an outer join, and an additional column called 'merge_indicator' is added to indicate the source of each row.
df_2022_Q1 = df_cleaned_q1.merge(daily_trips_q1.rename(columns = {'trip_id':'bike_rides_daily'}), on = 'Date', how = 'outer', indicator = 'merge_indicator')
```

```python
print(df_2022_Q1['_merge'].value_counts(dropna = False))
```

```python
plt.figure(figsize=(10,5))
plt.title('Temperature per trips')
plt.plot(df_2022_Q1['AvgTemp'])
plt.plot(df_2022_Q1['bike_rides_daily'], color = 'red')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 5))
ax.plot(df_2022_Q1['bike_rides_daily'],color = "red")
ax2 = ax.twinx()
ax2.plot(df_2022_Q1["AvgTemp"], color = "blue")
plt.show()
```

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_2022_Q1['bike_rides_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatures on the y-axis using second axis object
ax2.plot(df_2022_Q1['AvgTemp'], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips from Jan 2022 - April 2022', fontsize = 18)
plt.show()
```

To create a dual axis chart I used the object-oriented paradigm. I started by creating a figure and two axes objects using plt.subplots(). The two axes objects are necessary to plot on two different y-axes. I used the ax.plot() method to plot the trip counts against the date (or time). Next I used the ax1.twinx() method to create a second y-axis sharing the same x-axis. Then I used the ax2.plot() method to plot temperatures on the second y-axis.Finally I added titles & displayed the plot using plt.show().

Observation
---------------
There appears to be a positive correlation between temperature and the number of bike rides. As the temperature increases, the number of bike rides generally seems to increase as well.
During colder periods, especially in January and early February, the number of bike rides is relatively low. As the temperature warms up towards late February, March, and April, there is a noticeable increase in the number of daily bike rides.
There are some peaks in bike usage that correspond with higher temperature days. Conversely, there are dips in bike rides that align with colder days or possibly days with other unfavorable weather conditions.

Broader Analysis
---------------------
Planning and Resource Allocation - Understanding the correlation between temperature and bike usage can help in planning bike-sharing resources, such as station stocking and bike maintenance schedules. More bikes might be needed at stations during warmer periods when ridership increases.
Marketing Strategies: If there's a clear trend of increased ridership with warmer weather, marketing campaigns can be timed around the onset of warmer weather to maximize ridership.
Environmental and Public Health Impact: Encouraging biking during warmer months can contribute to reducing traffic congestion and emissions. Understanding seasonal usage patterns can also help in promoting biking as part of public health initiatives.

```python
df_cleaned_q1['started_at'] = pd.to_datetime(df_cleaned_q1['started_at'])
df_cleaned_q1['ended_at'] = pd.to_datetime(df_cleaned_q1['ended_at'])

# Changing the columns data into milisecond
df_cleaned_q1['started_at']= df_cleaned_q1['started_at'].astype(np.int64) // 10**6
df_cleaned_q1['ended_at']= df_cleaned_q1['ended_at'].astype(np.int64) // 10**6
```

```python
df_cleaned_q1['started_at'] = df_cleaned_q1['started_at'].astype(float)
df_cleaned_q1['ended_at'] = df_cleaned_q1['ended_at'].astype(float)

# Calculating the trip duration
df_cleaned_q1['tripduration'] = (df_cleaned_q1['ended_at'] - df_cleaned_q1['started_at'])/(1000 * 60)
# print(df_cleaned_q1['tripduration'])

df_cleaned_q1['log_tripduration'] = df_cleaned_q1['tripduration'].apply(lambda x: np.log(x + 1))

sns.set_style('darkgrid')
plt.figure(figsize=(9,5))
sns.displot(df_cleaned_q1['log_tripduration'], bins = 25)
plt.title('Distribution of Trip Duration', fontsize=16)
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(9,5))
sns.displot(df_cleaned_q1['log_tripduration'], bins = 25, kde=True)
plt.title('Distribution of Trip Duration', fontsize=16)
plt.show()
```
