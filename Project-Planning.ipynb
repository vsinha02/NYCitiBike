```python
# Importing Libraries

import pandas as pd
import numpy as np
import os
import requests
import json
from datetime import datetime
```

```python
# Set the path to the main directory containing subfolders
folderpath = '/Users/vineetasinha/Documents/workspace/NYCitiBike/Data-Q1'
print(f"Base directory path: {folderpath}")

contents = os.listdir(folderpath)
print("Contents of the directory:")
for item in contents:
    print(item)
    
# Create a list to store DataFrames
FirstQuater = []

# Iterate through each subfolder in the base directory
for subfolder in os.listdir(folderpath):
    subfolder_path = os.path.join(folderpath, subfolder)
    
    # Ensure the subfolder path is a directory
    if os.path.isdir(subfolder_path):
        
        # List all CSV files in the subfolder
        csv_files = [f for f in os.listdir(subfolder_path) if f.endswith('.csv')]
        
        # Read each CSV file and append to the list
        for file in csv_files:
            file_path = os.path.join(subfolder_path, file)
            df = pd.read_csv(file_path)
            FirstQuater.append(df)

# Concatenate all DataFrames into a single DataFrame
Bike_Q1 = pd.concat(FirstQuater, ignore_index=True)

# Display the first few rows of the combined DataFrame
Bike_Q1.head()
```

```python
# Secure all items in the response that correspond to TAVG

avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
# Get only the date field from all average temperature readings

data_temp = [item['date'] for item in avg_temps]
# Get the temperature from all average temperature readings

temps = [item['value'] for item in avg_temps]
```

```python
# Put the results in a dataframe

df_temps = pd.DataFrame()
# Get only date and case it to date time; convert temperature from tenths of celsius to normal celsius

df_temps['Date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in data_temp] 
df_temps['AvgTemp'] = [float(v)/10.0 for v in temps]

df_temps.tail()
```
