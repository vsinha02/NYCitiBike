```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
import requests
import json
from datetime import datetime
```

```python
# Reading the final dataset NYbike_trips_Q3.csv

filepath = '/Users/vineetasinha/Documents/workspace/NYCitiBike/NYbike_trips_Q2.csv'
df_NYTrips_Q2 = pd.read_csv(filepath)
```

```python
# Count missing values per column
missing_values_count = df_NYTrips_Q2.isnull().sum()

# Display count of missing values
print("Missing values count per column:")
print(missing_values_count)
```

```python
# Cleaning the missing values
df_cleaned_q2 = df_NYTrips_Q2.dropna(subset=['AvgTemp','end_station_name','end_station_id'])
```

```python
missing_values_count = df_cleaned_q2.isnull().sum()

# Display count of missing values
print("Missing values count per column:")
print(missing_values_count)
```

```python
# converting the 'Date' column in the df_NYTrips_Q1 DataFrame to a datetime format. It includes an additional parameter errors='coerce', which handles any errors that may arise during the conversion.
df_cleaned_q2['Date'] = pd.to_datetime(df_cleaned_q2['Date'], errors='coerce')
```

```python
# creating a new DataFrame named daily_trips_q1 that contains the count of bike trips per day for the second quarter (Q2).
daily_trips_q2 = pd.DataFrame(df_cleaned_q2.groupby(['Date'])['ride_id'].count()).reset_index()
```

```python
# rename a column in the daily_trips_q2 from ride_id to bike_rides_daily
daily_trips_q2.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)

# merging two DataFrames (df_NYTrips_Q2 and daily_trips_q2) on a common column, 'Date', while renaming a column from daily_trips_q1 in the process. The merge is performed using an outer join, and an additional column called 'merge_indicator' is added to indicate the source of each row.
df_2022_Q2 = df_cleaned_q2.merge(daily_trips_q2.rename(columns = {'trip_id':'bike_rides_daily'}), on = 'Date', how = 'outer', indicator = 'merge_indicator')
```

```python
print(df_2022_Q2['_merge'].value_counts(dropna = False))
```

```python
plt.figure(figsize=(10,5))
plt.title('Temperature per trips')
plt.plot(df_2022_Q2['AvgTemp'])
plt.plot(df_2022_Q2['bike_rides_daily'], color = 'red')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10, 5))
ax.plot(df_2022_Q2['bike_rides_daily'],color = "red")
ax2 = ax.twinx()
ax2.plot(df_2022_Q2["AvgTemp"], color = "blue")
plt.show()
```

```python
# Create figure and axis objects with subplots() and details

fig,ax = plt.subplots(figsize=(10, 5))

# Make a plot of the bike rides
ax.plot(df_2022_Q2['bike_rides_daily'],color = "navy")
ax.set_xlabel("Year 2022", fontsize = 14) # set x-axis label
ax.set_ylabel("Bike rides daily",color = "navy", fontsize = 14) # set y-axis label

# create the dual axis - initialize a twin object for two different y-axis on the sample plot
ax2 = ax.twinx()

# Make a plot with the temperatures on the y-axis using second axis object
ax2.plot(df_2022_Q2['AvgTemp'], color = "red")
ax2.set_ylabel("Average temperatures", color = "red",fontsize=14)
plt.title('Temperature and trips from May 2022 - August 2022', fontsize = 18)
plt.show()
```

To create a dual axis chart I used the object-oriented paradigm. I started by creating a figure and two axes objects using plt.subplots(). The two axes objects are necessary to plot on two different y-axes. I used the ax.plot() method to plot the trip counts against the date (or time). Next I used the ax1.twinx() method to create a second y-axis sharing the same x-axis. Then I used the ax2.plot() method to plot temperatures on the second y-axis.Finally I added titles & displayed the plot using plt.show().

Observation
---------------

The number of daily bike rides is consistently high, often exceeding 100,000 rides per day. This suggests that May through August are peak months for bike-sharing services.
The consistent increase in bike rides aligns with the warmer temperatures during these months.
Average temperatures in this period range mostly between 20°C and 30°C, which are generally favorable for outdoor activities. The data shows that even as temperatures rise above 25°C, bike usage remains strong, indicating that such temperatures are within the comfortable range for most users.
Unlike the first part of the year, where bike ridership was more variable, the ridership in these months appears to plateau at a high level, reflecting the stable, warm weather. This stability might indicate that external factors (e.g., holidays, events) play a lesser role than weather in determining ridership during these months.

Broader Analysis
---------------------
Resource Planning - The stable and high ridership during these months suggests that bike stations should be well-stocked and maintained to handle the increased demand. This can also be used to plan for potential expansions of service in similar periods.
Marketing efforts should focus on the early summer months to capitalize on the peak in ridership. This could include promotions, partnerships, and campaigns encouraging biking as a mode of transport.
Environmental Impact - With high ridership during the warmer months, the environmental benefits of cycling, such as reduced carbon emissions and less traffic congestion, are amplified. Analyzing these trends can help justify investments in cycling infrastructure and support sustainability initiatives.

```python
df_cleaned_q2['started_at'] = pd.to_datetime(df_cleaned_q2['started_at'])
df_cleaned_q2['ended_at'] = pd.to_datetime(df_cleaned_q2['ended_at'])

# Changing the columns data into milisecond
df_cleaned_q2['started_at']= df_cleaned_q2['started_at'].astype(np.int64) // 10**6
df_cleaned_q2['ended_at']= df_cleaned_q2['ended_at'].astype(np.int64) // 10**6
```

```python

df_cleaned_q2['started_at'] = df_cleaned_q2['started_at'].astype(float)
df_cleaned_q2['ended_at'] = df_cleaned_q2['ended_at'].astype(float)

# Calculating the trip duration
df_cleaned_q2['tripduration'] = (df_cleaned_q2['ended_at'] - df_cleaned_q2['started_at'])/(1000 * 60)
print(df_cleaned_q2['tripduration'])

df_cleaned_q2['log_tripduration'] = df_cleaned_q2['tripduration'].apply(lambda x: np.log(x + 1))

sns.set_style('darkgrid')
plt.figure(figsize=(9,5))
sns.displot(df_cleaned_q2['log_tripduration'], bins = 25)
plt.title('Distribution of Trip Duration', fontsize=16)
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(9,5))
sns.displot(df_cleaned_q2['log_tripduration'], bins = 25, kde=True)
plt.title('Distribution of Trip Duration', fontsize=16)
plt.show()
```
